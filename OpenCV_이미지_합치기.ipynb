```python
import cv2
import matplotlib.pyplot as plt

img1 = cv2.imread("image_1.jpg")
img2 = cv2.imread("image_2.png")

# Saturation 연산
result = cv2.add(img1,img2)
plt.imshow(cv2.cvtColor(result,cv2.COLOR_BGR2RGB))
plt.show()

# Modulo 연산
result = img1 + img2
plt.imshow(cv2.cvtColor(result,cv2.COLOR_BGR2RGB))
plt.show()
```

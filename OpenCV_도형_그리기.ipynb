### 직선 그리기

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = np.full((512,512,3),255,np.uint8)
img = cv2.line(img,(0,0),(255,255),(255,0,0),2)
plt.imshow(img)
plt.show()
```

### 사각형 그리기

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = np.full((512,512,3),255,np.uint8)
img = cv2.rectangle(img,(20,20),(255,255),(255,0,0),-1)
plt.imshow(img)
plt.show()
```

### 원 그리기

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = np.full((512,512,3),255,np.uint8)
inmg = cv2.circle(img,(255,255),30,(255,0,0),3)

plt.imshow(img)
plt.show()
```

### 다각형 그리기

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = np.full((512,512,3),255,np.uint8)
points = np.array([[5,5],[128,256],[483,444],[400,150]])
img = cv2.polylines(img,[points],True,(0,255,0),4)

plt.imshow(img)
plt.show()
```

### 텍스트 그리기

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = np.full((512,512,3),255,np.uint8)
img = cv2.putText(img,"Hello World",(0,200),cv2.FONT_ITALIC,2,(255,0,0))

plt.imshow(img)
plt.show()
```

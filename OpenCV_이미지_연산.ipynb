### OpenCV를 활용한 이미지 크기 및 픽셀 정보 확인

```python
import cv2
img = cv2.imread("cat.jpg")

# 픽셀 수 및 이미지 크기 확인
print(img.shape) # 이미지 크기
print(img.size) # 픽셀 수

# 이미지 Numpy 객체의 특정 픽셀(100 x 100)
px = img[100,100]

# B,G,R 순으로 출력
# 단, Gray Scale인 경우에는 B,G,R로 구분되지 않는다.
print(px) # [B,G,R]

# R값만 출력하기
print(px[2])
```

### OpenCV를 활용한 특정 범위 픽셀 변경

```python
import cv2
import matplotlib.pyplot as plt
import time
img = cv2.imread("cat.jpg")

start_time = time.time()
for i in range(100):
  for j in range(100):
    img[i,j] = [255,255,255] # 흰색으로 변경
print("--- %s seconds ---" % (time.time() - start_time))
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()

start_time = time.time()
img[0:100, 0:100] = [0,0,0] # 검은색으로 변경
print("--- %s seconds ---" % (time.time() - start_time))
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()
```

### OpenCV를 활용한 ROI(Region of Interest: 관심 있는 영역)추출

```python

import cv2
import matplotlib.pyplot as plt

img = cv2.imread('cat.jpg')

# Numpy 슬라이싱: ROI 처리 가능
roi = img[200:350, 50:200]
plt.imshow(cv2.cvtColor(roi,cv2.COLOR_BGR2RGB))
plt.show()

# ROI 단위로 이미지 복사하기
img[:150, :150] = roi
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()
```

### OpenCV를 활용한 픽셀별 색상 다루기

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('cat.jpg')
img[:,:,1] = 0  # 모든 픽셀의 G값을 0으로 바꾼다, 즉 모든 픽셀에 대해서 G값이 사라진다.

plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()
```
